```python
import matplotlib.pyplot as plt
import csv

plt.style.use('seaborn-bright')

fig, ax = plt.subplots()

with open('data/kiel_bevoelkerung_familienstand.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=';')
    
    # create empty lists
    jahr = []
    ledig = []
    verheiratet = []
    
    for row in readCSV:
        #print(row)
        #print(row[4],row[5],row[6],row[7],row[8])
        
        a = row[4]
        b = row[5]
        c = row[6]
        #d = row[7]
        #e = row[8]
        
        jahr.append(a)
        ledig.append(b)
        verheiratet.append(c)

xlabels = [int(i) for i in jahr]

ax.set_title("Kiel - Familienstand", size="x-large")
ax.set_ylabel("in Tsd", size="x-large")
ax.set_xlabel("Jahr", size="x-large")

ax.plot(ledig, "r*-", markersize=6, linewidth=1, color='g', label="ledig")
ax.plot(verheiratet, "r*-", markersize=6, linewidth=1, color='r', label="verheiratet")

ax.legend(loc=(0.25, 0.75))

ax.set_xticks(range(len(jahr)))
ax.set_xticklabels(xlabels, rotation='vertical')

plt.show()
```
